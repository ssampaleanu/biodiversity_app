```python
import pandas as pd
microbe_counts = pd.read_sql_table('samples', engine, index_col = 'otu_id')
microbe_counts.head()
```

```python
otu_ids = []
sample_values = []

for row in microbe_counts.iterrows():
    search_col = "BB_940"
    if row[search_col].value != 0:
        sample_values.append(row[search_col].value)
        otu_ids.append(row.index)
        
otu_ids
```
